```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    accuracy_score, classification_report, precision_score,
    recall_score, f1_score, confusion_matrix, ConfusionMatrixDisplay,
    roc_curve, auc
)
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier  # Yapay Sinir A (MLP) ekleniyor


# Veri Ykleme ilemi
# Bu veri seti, evrimii alveri yapan kiilerin davranlarn ve alveri yapp yapmadklarn ieriyor.
df = pd.read_csv("online_shoppers_intention.csv")


# Makine renmesi algoritmalar sadece saysal verilerle alr. Bu yzden:
# Kategorik Veriler saysal verilere dntrld ve
# Makine renmesi algoritmalarnn kullanlabilecei forma getirilmi oldu.
le = LabelEncoder() # LabelEncoder snfndan "le" adl bir nesne yani dntrc oluturuldu.

# String -> Integer
df['Month'] = le.fit_transform(df['Month']) # fit ifadesi veri zerinde inceleme yapar(esiz deerleri bulur)
df['VisitorType'] = le.fit_transform(df['VisitorType']) # transform ise rendiklerine gre veriyi dntrr.
# Yukardaki month ve visitorType verileri string veri trnde olduklar iin fit ve transorm yntemleri uyguland.


# Boolean -> Integer(1/0)
# Aadaki weekend ve revenue verileri ise boolean(t/f) veri trnde olduklar iin direkt integer'a(1/0) dnmleri saland.
df['Weekend'] = df['Weekend'].astype(int)
df['Revenue'] = df['Revenue'].astype(int)
# Weekend ve Revenue stunlar da True/False -> 0/1 olarak dntrld.


# zellikler ve Hedef
X = df.drop('Revenue', axis=1) # drop komutu Revenue stununu veriden karr nk
                                       # Revenue bizim tahmin etmek istediimiz ey yani hedef stunu.

y = df['Revenue'] # kt yani modelin tahmin etmeye alt ey.


# Normalizasyon
scaler = MinMaxScaler() # Tm saysal deerleri 0 ile 1 arasnda sktrr.
X_scaled = scaler.fit_transform(X)

# Veriyi Ayrma ilemi modelin grd veriyi ezberlemeyip genelleme yapmas iin nemlidir.
# Eitim verisi ile model renir.
# Test verisi ile modelin baars kontrol edilir.
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y # Verinin %20si test iin ayrlacak.
)
  # random_state ifadesi ayn sonular almak iin sabit rastgelelik salar.
  # stratify ifadesi Eitim ve test setlerinde, baml deikenin (y) snf dalmnn korunmasn salar.
  # zellikle snf dengesizlii varsa ok nemlidir.


# 7 farkl makine renmesi algoritmasn ieren dictionary tanm.
models = {
    "KNN": KNeighborsClassifier(n_neighbors=5), # Eitim verisindeki en yakn 5 komuya bakar, en ok hangi snf varsa onu tahmin eder.

    "Decision Tree": DecisionTreeClassifier(random_state=42),

    "SVM": SVC(probability=True), # probability=True ifadesi ile Tahmin sonucu sadece 0/1 deil, ayn zamanda olaslk verebilmesi salanr.
                                  # SVM Her iki snfa da en uzakta kalan izgiyi yani en gvenli ayrc snr bulmaya alr.

    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42), # n_estimators=100: 100 tane karar aac kurar. Her aa ayr karar verir, ounluk ne diyorsa o olur

    "Naive Bayes": GaussianNB(), # Olaslk temelli alr. zellikler birbirinden bamszm gibi varsayar.

    "Logistic Regression": LogisticRegression(max_iter=1000), # Veriyi sigmoid fonksiyonuyla 0-1 aras deerlere sktrarak snflandrma yapar.
                                                              # max_iter=1000 ise zor problemler iin daha fazla deneme hakk verir

    "Neural Network": MLPClassifier(hidden_layer_sizes=(100,), max_iter=1000, random_state=42)  # Yapay Sinir A (MLP).
    # Girdi alr - zerinde matematiksel ilem yapar - Sonraki katmana iletir

    # hidden_layer_sizes=(100,) 100 nronlu bir gizli katman kullanr.
    # max_iter=1000: ile de renmesi iin 1000 adm (iterasyon) tannr.
}

# Sonular tutmak iin
metrics = {
    "Model": [],
    "Accuracy": [],  # Doruluk
    "Precision": [], # Kesinlik
    "Recall": [],    # Duyarllk
    "F1 Score": []   # Kesinlik ve Duyarlln dengeli ortalamas
}


roc_data = {} # ROC erisi izmek iin gerekli olan FPR, TPR ve AUC deerlerini saklayacamz bo bir szlk.
# fpr - false positive rate (yanl alarmlar)
# tpr - true positive rate (doru alarmlar)
# roc_auc - eri altnda kalan alan (modelin ne kadar iyi ayrlabildiini gsterir)


# Eitim ve Deerlendirme
for name, model in models.items(): # Tm modeller dng yardmyla teker teker eitilip test ediliyor.
    model.fit(X_train, y_train)    # Modeli eitmek amacyla modeli inceliyor.
    # X_train : Modelin renecei eyin girdisidir.
    # Y_train : Modelin tahmin etmeye alt eydir.

    y_pred = model.predict(X_test) # Eitim tamamlandktan sonra test verisiyle tahmin yapar.


    # Bu satrlarda model isimleri metrics szlnn "Model" listesine eklenr.
    # average='macro' ifadesi her snfn skorunu ayr ayr hesaplar ve bu skorlarn aritmetik ortalamasn alr. Her snfa eit nem verir.
    metrics["Model"].append(name)
    metrics["Accuracy"].append(accuracy_score(y_test, y_pred))
    metrics["Precision"].append(precision_score(y_test, y_pred, average='macro'))
    metrics["Recall"].append(recall_score(y_test, y_pred, average='macro'))
    metrics["F1 Score"].append(f1_score(y_test, y_pred, average='macro'))


    # Confusion Matrix (Karmaklk Matrisi)
    cm = confusion_matrix(y_test, y_pred) # Modelin tahminlerinin doru ve yanl olduu durumlar bir 2x2 tablo(matris) olarak verir.

    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['No Purchase', 'Purchase'])  # Etiketleri zelletirir.
    # Bu snf, confusion matrix'i grsel olarak izmeye yarar.
    # 'No Purchase', 'Purchase' ifadeleri ile 0 ve 1 snfna anlaml isimler veriliyor.

    disp.plot(cmap='Blues') # Matrisin grselini oluturur ve grselde mavi renk tonlar kullanlr .
    plt.title(f"{name} - Confusion Matrix")
    plt.grid(False) # Klavuz izgileri kapatlr.
    plt.show()


    # ROC erisi, bir modelin snflandrmadaki baarsn olas tm eik (threshold) deerleri iin gsteren bir grafik trdr.

    if hasattr(model, "predict_proba"): # Modelin predict_proba adnda bir fonksiyonu var m diye kontrol eder.
        y_probs = model.predict_proba(X_test)[:, 1] # Modelin her test rnei iin [negatif snf olasl, pozitif snf olasl] dndrmesini salar.

    elif hasattr(model, "decision_function"):   # decision_function() denen baka bir yntemle, karar snrna olan uzaklk bulunur.
        y_probs = model.decision_function(X_test) # y_probs, her test rnei iin pozitif snfa (Revenue = 1) ait olma ihtimallerinin listesidir.
        y_probs = (y_probs - y_probs.min()) / (y_probs.max() - y_probs.min()) # ROC izmek iin normalize edilir.
    else:
        y_probs = np.zeros_like(y_pred) # y_pred ile ayn boyutta ama btn elemanlar sfr olan bir NumPy dizisi (array) oluturur.
                                        # Bu,olaslk deeri salayamayan modeller iin bir yedek (default) deer retir.

    fpr, tpr, _ = roc_curve(y_test, y_probs) # Bu fonksiyon, farkl eik deerlerinde True Positive Rate ve False Positive Rate hesaplar.
    # sadece fpr ve tpr ROC izimi iin yeterli, eikler izim iin kullanlmayacak. Bu yzden _ kullanld.
    roc_auc = auc(fpr, tpr) # ROC erisinin altnda kalan alan (AUC) hesaplar.
    roc_data[name] = (fpr, tpr, roc_auc) # her model iin hesaplanan fpr tpr ve auc deerlerini bir szlkte sakluyor.

# Metrikleri DataFramee evir
metrics_df = pd.DataFrame(metrics)

# Skorlar Grselletir
# Veriyi grafikte kolay izilebilecek uzun formata evirir.
melted_df = metrics_df.melt(id_vars='Model', var_name='Metrik', value_name='Skor')

plt.figure(figsize=(14, 6))
sns.barplot(data=melted_df, x='Model', y='Skor', hue='Metrik', palette='bright')

plt.title('Modellerin Accuracy, Precision, Recall ve F1 Skorlar')
plt.ylim(0, 1) # Y ekseninin minimum ve maksimum snrlarn belirler
plt.ylabel('Skor')
plt.grid(axis='y', linestyle='--', alpha=0.7) # alpha ile effaflk derecesi ayarlanr

# bbox_to_anchor=(1.05, 1) - Kutuyu grafiin dna, sa st keye tar.
# loc='upper left' - Kutunun konumunu sol stte sabitler ama (1.05, 1) konumunda.
plt.legend(title='Metrik', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout() # Grafik elemanlarnn (balk, etiketler, ubuklar, yazlar) birbirine akmasn engeller.
plt.show()

# ROC Erisi izimi
plt.figure(figsize=(10, 6))
for name, (fpr, tpr, roc_auc) in roc_data.items():
    plt.plot(fpr, tpr, label=f'{name} (AUC = {roc_auc:.2f})')

plt.plot([0, 1], [0, 1], 'k--', label='ans (AUC = 0.5)') # Bu, rastgele tahmin yapan modelin ROC erisidir (kegen izgi).
# ROC erisi grafiine "ans izgisi" ekler. Bu izgi, rastgele tahmin yapan bir modelin performansn temsil eder.
# Bu izgi, TPR = FPR olan tm noktalar gsterir  yani model doru tahmin yapamyor, tamamen rastgele karar veriyor demektir.
# Rastgele tahmin yapan bir modelin ROC AUC skoru her zaman 0.5 olur

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Erisi')
plt.legend(loc='lower right') # Bu komut, grafikte kullanlan renklerin veya izgilerin neyi temsil ettiini gsteren kk bir kutu ekler.
plt.grid(True)
plt.show()
```
